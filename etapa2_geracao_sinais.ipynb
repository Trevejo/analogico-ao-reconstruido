# Etapa 2: Geração de Sinais Periódicos

**Objetivo:** Construir e visualizar três tipos clássicos de sinais periódicos (onda quadrada, triangular, dente de serra).

**Tarefas:**
* Gerar os três sinais com a mesma frequência fundamental (escolha uma frequência fundamental razoável, por exemplo, 1 Hz).
* Plotar os gráficos no domínio do tempo (eixo x: tempo, eixo y: amplitude). Certifique-se de que cada sinal mostre pelo menos 2 ciclos completos.
* Identificar as características visuais e temporais de cada onda (discontinuidades, suavidade, simetria) em um texto descritivo ou tabela.
* Escolher um sinal para seguir nas próximas etapas, justificando a escolha.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

## Geração dos Sinais

```python
# Parâmetros dos sinais
f_fundamental = 1  # Frequência fundamental em Hz
T = 1 / f_fundamental # Período
amplitude = 1
n_ciclos = 2 # Número de ciclos para visualização
t = np.linspace(0, n_ciclos * T, 500, endpoint=False) # Vetor de tempo
```

```python
# Onda Quadrada
onda_quadrada = amplitude * signal.square(2 * np.pi * f_fundamental * t)

# Onda Triangular
# A função sawtooth com width=0.5 gera uma onda triangular
onda_triangular = amplitude * signal.sawtooth(2 * np.pi * f_fundamental * t, width=0.5)

# Onda Dente de Serra
# A função sawtooth com width=1 (ou omitido) gera uma dente de serra ascendente
onda_dente_serra = amplitude * signal.sawtooth(2 * np.pi * f_fundamental * t)
```

## Visualização dos Sinais

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, onda_quadrada)
plt.title('Onda Quadrada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, onda_triangular)
plt.title('Onda Triangular')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, onda_dente_serra)
plt.title('Onda Dente de Serra')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Comparação Descritiva dos Sinais

| Característica     | Onda Quadrada                                  | Onda Triangular                                     | Onda Dente de Serra                               |
|--------------------|------------------------------------------------|-----------------------------------------------------|---------------------------------------------------|
| **Discontinuidades** | Presentes nas transições abruptas de nível.    | Ausentes, a onda é contínua.                        | Presentes no final de cada ciclo (salto abrupto). |
| **Suavidade**      | Não suave, possui cantos vivos.                | Suave, com transições lineares.                     | Não suave nas transições de ciclo.                |
| **Simetria**       | Simetria par (se centrada na origem e olhando meio período) e simetria ímpar (se deslocada). Apresenta simetria de meia onda. | Simetria par em torno dos picos e vales. Apresenta simetria de meia onda. | Assimétrica.                                      |
| **Variação**       | Varia entre dois níveis de amplitude (-A e +A). | Varia linearmente entre -A e +A.                   | Varia linearmente de -A a +A e depois salta para -A.|
| **Derivada**       | Impulsos nas discontinuidades, zero no resto.   | Onda quadrada.                                      | Constante na maior parte do período, impulso na descontinuidade. |

## Escolha do Sinal para Próximas Etapas

**Sinal Escolhido:** Onda Quadrada.

**Justificativa:** A onda quadrada é uma excelente escolha para ilustrar os conceitos da Série de Fourier e os efeitos da digitalização devido às suas características distintas:
1.  **Rica em Harmônicas:** Possui um espectro de frequência rico, contendo apenas harmônicas ímpares da frequência fundamental. Isso torna a análise de Fourier mais interessante e visualmente clara, permitindo observar como a soma de senoides (ou cossenos) aproxima as transições abruptas.
2.  **Discontinuidades Claras:** As discontinuidades da onda quadrada são um bom exemplo do fenômeno de Gibbs quando a série é truncada, o que é um ponto importante na análise de Fourier.
3.  **Representação Comum:** É um sinal fundamental em eletrônica digital e sistemas de comunicação, tornando sua análise relevante em diversos contextos práticos.
4.  **Desafio para Reconstrução:** A necessidade de muitas harmônicas para uma boa aproximação das bordas agudas ilustra bem os conceitos de largura de banda e a importância das componentes de alta frequência.
